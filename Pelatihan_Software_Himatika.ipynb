# Pemodelan Data

```python
asuransi_mobil43_dilatih = asuransi_mobil43.drop(['KLAIM'],axis=1)
asuransi_mobil43_target = asuransi_mobil43['KLAIM']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(asuransi_mobil43_dilatih,asuransi_mobil43_target,test_size=0.2)
```

```python
from sklearn.svm import SVC
model_svm = SVC()
model_svm.fit(X_train,y_train)
model_svm.score(X_test,y_test)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=5, shuffle=True, random_state=0)
```

```python
model = SVC()
score = cross_val_score(model, asuransi_mobil43_dilatih,asuransi_mobil43_target,cv=k_fold,n_jobs=1,scoring='accuracy')
print(score)
```
